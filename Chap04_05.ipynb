## 4.5 데이터의 관계를 설명하는 선형 회귀 함수의 시각적 이해

```python
import numpy as np

from sklearn.metrics import mean_squared_error as mse

x = np.array([1, 4.5, 9, 10, 13])
y = np.array([0, 0.2, 2.5, 5.4, 7.3])

w_list = np.arange(1.0, 0.2, -0.1)  # 1.0에서 0.2까지 0.1씩 줄어드는 리스트
for w in w_list :
    y_hat = w * x
    w_mse = mse(y_hat, y)
    print(f'w = {w:.1f}, 평균 제곱 오차 : {w_mse:.2f}')
```

```python
# 도전문제 4.1_1번
import matplotlib.pyplot as plt

def w_mse(y_hat, y) :
    return mse(y_hat, y)

w_x = []

w_list = np.arange(1.0, 0.0, -0.1)
for w in w_list :
    y_hat = w * x
    w_x.append(w_mse(y_hat, y))
    
plt.plot(w_list, w_x)
plt.xlabel('w')
plt.ylabel('mse')
plt.title('MSE optimization graph')
```
