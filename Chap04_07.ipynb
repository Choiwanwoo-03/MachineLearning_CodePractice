### 4.7 오차로 가설을 평가하고 좋은 가설 찾기

```python
import matplotlib.pyplot as plt
import pandas as pd

data_home = 'https://github.com/dknife/ML2nd/raw/main/data/'
lin_data = pd.read_csv(data_home + 'pollution.csv')

def h(x, w, b) :
    return w * x + b

w, b = -3, 6
x = lin_data['input'].to_numpy()
y = lin_data['pollution'].to_numpy()
y_pred = h(x, w, b)
error = (y_pred - y)
error
```

```python
learning_rate = 0.003
w = w - learning_rate * (error * x).sum()
b = b - learning_rate * error.sum()
```

```python
lin_data.plot(kind = 'scatter', x = 'input', y = 'pollution')
x0, x1 = 0.0, 1.0
plt.plot([x0, x1]), [h(x0, w, b), h(x1, w, b)]
```

```python
for i in range(100) :
    y_pred = h(x, w, b)
    error = (y_pred - y)
    error
    w = w - learning_rate * (error * x).sum()
    b = b - learning_rate * error.sum()
    
    if i % 20 == 0 :
        lin_data.plot(kind = 'scatter', x = 'input', y = 'pollution')
        x0, x1 = 0.0, 1.0
        plt.plot([x0, x1], [h(x0, w, b), h(x1, w, b)])
```
